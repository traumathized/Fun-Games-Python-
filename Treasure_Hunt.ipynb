```python
import random

wins = 0
losses = 0

print("🏴‍☠️ Welcome to Treasure Hunt Game!")

while True:
    treasure_door = random.randint(1, 5)

    choice = input("Choose a door (1-5) or type 'quit': ")

    if choice == "quit":
        print("\nGame Over!")
        print("Wins:", wins)
        print("Losses:", losses)
        break

    # Convert input to number
    choice = int(choice)

    if choice >= 1 and choice <= 5:
        if choice == treasure_door:
            print("🎉 You found the treasure!")
            wins += 1
        else:
            print("💀 Trap! You lost!")
            print("The treasure was behind door", treasure_door)
            losses += 1
    else:
        print("❌ Invalid door number! Choose between 1 and 5.")

    print("Score → Wins:", wins, "| Losses:", losses)
    print()
```
