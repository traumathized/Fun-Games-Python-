```python
import random

print("🎲 Welcome to Dice Roll Challenge!")
print("You are playing against the Computer.")
print("Type 'roll' to roll the dice or 'quit' to exit.\n")

user_score = 0
computer_score = 0
round_number = 0

while True:
    choice = input("Enter roll or quit: ").lower()

    if choice == "quit":
        print("\n🎮 Game Over!")
        print("Final Score:")
        print("You:", user_score)
        print("Computer:", computer_score)

        if user_score > computer_score:
            print("🏆 You won the game!")
        elif computer_score > user_score:
            print("💻 Computer won the game!")
        else:
            print("🤝 It's a tie!")
        break

    elif choice == "roll":
        round_number += 1
        print(f"\n--- Round {round_number} ---")

        user_dice = random.randint(1, 6)
        computer_dice = random.randint(1, 6)

        print("🎲 You rolled:", user_dice)
        print("🎲 Computer rolled:", computer_dice)

        if user_dice > computer_dice:
            print("🎉 You win this round!")
            user_score += 1
        elif computer_dice > user_dice:
            print("💻 Computer wins this round!")
            computer_score += 1
        else:
            print("🤝 This round is a tie!")

        print("Score → You:", user_score, "| Computer:", computer_score, "\n")

    else:
        print("❌ Invalid input! Please type 'roll' or 'quit'.\n")
```
